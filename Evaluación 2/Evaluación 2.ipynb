```python
#Ahora vamos con la ET0 de Valiantzas 4
#La ecuación es: ET0 = 0.051(1 - α)Rs(Tmean + 9.5)^0.5 - 2.4(Rs/Ra)^2 + 0.048(Tmean + 20)(1 - RH/100)(0.5 + 0.536u2) + 0.00012z
#Donde α es el albedo (Albedo_Avg)
#Ra es la radiación solar en la parte alta de la atmósfer
#u2 es la velocidad del viento a 2 m de altura (WS_ms_S_WVT)
#z es la altura sobre el nivel mar

#Para sacar Ra usamos la siguiente ecuación:
#Ra = ((24*60)/π)Gsc*dr[ωs*Sen(φ)Sen(δ) + Cos(φ)Cos(δ)Sen(ωs)]
#Donde Gsc es la constante solar, y vale 0.0820 MJ m-2 min-1
#dr es la inversa relativa de la distancia Tierra-Sol (dr = 1 + 0.033Cos((2π/365)*J)) --- (J es el número del día en el año)
#ωs es el áungulo de la hora del atardecer en radianes (ωs = Arccos[-Tan(φ)Tan(δ)])
#φ es la latitud en radianes (0.5047 rad)
#δ es la decimación solar (δ = 0.409Sen((2π/365)*J - 1.39)) --- (J es el número del día en el año)

#En esta parte primero conseguiremos la radiación solar y luego trabajaremos el resto
#Así que lo primero que haremos será definir las constantes
π=math.pi
φ=0.5047
z=101
Gsc=0.0820

#Luego sacaremos δ y dr de cada día
#Primero δ
dfδ=[0] * 365
j=1

for i in range(0,365):
    dfδ[i]=0.409*math.sin(((2*π)/365)*j - 1.39)
    j=j+1

#Y ahora dr
dfdr=[0] * 365
j=1

for i in range(0,365):
    dfdr[i]=1 + 0.033*math.cos(((2*π)/365)*j)
    j=j+1

#Y con eso conseguiremos ωs de cada día
dfωs=[0] * 365

for i in range(0,365):
    dfωs[i]=math.acos(-math.tan(φ)*math.tan(dfδ[i]))

#Y finalmente conseguimos con todo esto la radiación solar (Ra) diaria
dfRA=[0] * 365

for i in range(0,365):
    dfRA[i]=((24*60)/π)*Gsc*dfdr[i]*(dfωs[i]*math.sin(φ)*math.sin(dfδ[i]) + math.cos(φ)*math.cos(dfδ[i])*math.sin(dfωs[i]))
```

```python
#Ahora, haremos un arreglo que contenga las horas que hay en un mes (720)
#Cabe aclarar, que tomaremos la duración de un mes como 30 días
dfHM=[0] * 720

#Y representaremos las horas del día (24), cada día del mes (30)
#De esta forma tendríamos una sucesión que se repetirá 30 veces
for i in range(0,720,24):
    dfHM[i]="0:00 - 1:00"
    dfHM[i+1]="1:00 - 2:00"
    dfHM[i+2]="2:00 - 3:00"
    dfHM[i+3]="3:00 - 4:00"
    dfHM[i+4]="4:00 - 5:00"
    dfHM[i+5]="5:00 - 6:00"
    dfHM[i+6]="6:00 - 7:00"
    dfHM[i+7]="7:00 - 8:00"
    dfHM[i+8]="8:00 - 9:00"
    dfHM[i+9]="9:00 - 10:00"
    dfHM[i+10]="10:00 - 11:00"
    dfHM[i+11]="11:00 - 12:00"
    dfHM[i+12]="12:00 - 13:00"
    dfHM[i+13]="13:00 - 14:00"
    dfHM[i+14]="14:00 - 15:00"
    dfHM[i+15]="15:00 - 16:00"
    dfHM[i+16]="16:00 - 17:00"
    dfHM[i+17]="17:00 - 18:00"
    dfHM[i+18]="18:00 - 19:00"
    dfHM[i+19]="19:00 - 20:00"
    dfHM[i+20]="20:00 - 21:00"
    dfHM[i+21]="21:00 - 22:00"
    dfHM[i+22]="22:00 - 23:00"
    dfHM[i+23]="23:00 - 24:00"

#Y haremos otra donde aparezca el número de los días
#(Se repetirá 30 veces para que se pueda ver al final el día y la hora de este)
dfDHM=[0] * 720
j=1

for i in range(0,720,24):
    for k in range(0,24):
        dfDHM[i+k]=j
    j=j+1
```

```python
#Ahora elegiremos un intervalo de datos de Rn-G, λET y H para trabajar
#Como debemos agarrar un mes, utilizaremos del día 91 al 120 (en teoría correspondería al mes de abril)
#(Esto corresponde del elemento 2016 al 2736)
dfRNGHI=dfRNGH[2016:2736]
dfλETHI=dfλETH[2016:2736]
dfHHI=dfHH[2016:2736]
```
