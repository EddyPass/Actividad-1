```python
#Invocar Pandas
import pandas as pd
import numpy as np
```

```python
#Leemos los datos, saltandonos hasta el primero de noviembre
df0 = pd.read_csv("vid18_180219.dat",skiprows=range(1, 24936))
df0
```

```python
#Agarramos las columnas quer nos interesan
df1 = df0[["TIMESTAMP","AirTC_Avg"]]
df1
```

```python
#Declararlo como una función de Panda
df = pd.DataFrame(df1)
```

```python
#Promediamos las horas, para ello tomamos seis filas y las promediamos, y así sucesivamente hasta terminar
#Lo ponemos todo en un arreglo
#Haremos uso de un loop
j=0
df3=[0] * 2621

for i in range(0, 15720, 6):
    df2=df1[i:i+6].mean()
    i=i+5
    df3[j]=df2
    j=j+1

#Como el loop no incluyó el promedio de los últimos cuatro números (porque no hacían una hora), se saca a parte y se agrega al arreglo
df2=df1[15721:15724].mean()
df3[2620]=df2
```

```python
#Y lo declaramos como una función de Pandas
df4 = pd.DataFrame(df3)
df4
```

```python
#Comparamos cada elemento para sacar el UF y lo metemos en otro arreglo

df5=[0] * 2621

for k in range(0, 2621):
    if df3[k].item() < 1.4:
        df5[k]=0

    elif 1.5 < df3[k].item() and df3[k].item() < 2.4:
        df5[k]=0.5

    elif 2.5 < df3[k].item() and df3[k].item() < 9.1:
        df5[k]=1

    elif 9.2 < df3[k].item() and df3[k].item() < 12.4:
        df5[k]=0.5

    elif 12.5 < df3[k].item() and df3[k].item() < 15.9:
        df5[k]=0

    elif 16 < df3[k].item() and df3[k].item() < 18:
        df5[k]=-0.5

    elif df3[k].item() > 18.0:
        df5[k]=-1
```

```python
#Y lo declaramos como una función de Pandas
df6 = pd.DataFrame(df5)
df6
```

```python
#Ahora sacamos la de cada día
#Sabemos que cada 24 intervalos de la tabla anterior equivalen a un día, así que hacemos un loop que saque el UF24 dividiendo la tabla anterior en intervalos y sumando los datos en ella
#El último día al estar incompleto se sacará a parte
#Metemos los datos en otro arreglo mientras vayamos sacando todo
u=0
df8=[0] * 111

for y in range(0, 2616, 24):
    df7=sum(df5[y:y+23])
    df8[u]=df7
    u=u+1

#Como el loop no incluyó el promedio de los últimos cuatro números (porque no hacían una hora), se saca a parte y se agrega al arreglo
df7=sum(df5[2617:2621])
df8[110]=df7
```
