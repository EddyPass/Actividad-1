```python
#Comparamos cada elemento para sacar el UF y lo metemos en otro arreglo

df5=[0] * 2621

for k in range(0, 2621):
    if df3[k].item() < 1.5:
        df5[k]=0

    elif 1.5 <= df3[k].item() and df3[k].item() < 2.5:
        df5[k]=0.5

    elif 2.5 <= df3[k].item() and df3[k].item() < 9.2:
        df5[k]=1

    elif 9.2 <= df3[k].item() and df3[k].item() < 12.5:
        df5[k]=0.5

    elif 12.5 <= df3[k].item() and df3[k].item() < 16:
        df5[k]=0

    elif 16 <= df3[k].item() and df3[k].item() < 18:
        df5[k]=-0.5

    elif df3[k].item() >= 18.0:
        df5[k]=-1
```
